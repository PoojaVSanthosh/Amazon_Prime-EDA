## 3. ***Data Wrangling***

Data Wrangling or Data Munging is the process of cleaning, transforming, and organizing raw data into a usable format for analysis. There are so many steps in data wrangling.

### Data Wrangling Code

Step 1 - Merging Dataset

Now, Have to merge credits and titles dataset for further analysis. Here I have used common *column id* for merging the both datasets

````python
prime=pd.merge(data1, data2, on='id')
print(prime.shape)
prime.head()
````

````python
print(prime.shape)
print(prime.columns)
````

````python
prime.info()
````

````python
prime.describe()   # it gives statistical information
````

````python
#duplicates and missing values in dataset prime
print ('Number of Duplicates:', prime.duplicated().sum())
print(prime.isnull().sum())
````

````python
#Unique values in Prime
prime.nunique()
````

**Step 3 - Droping columns and filling with appropirate Columns values**

Here, Some of the columns which is not be needed. So, we can drop these columns

Here, I'm droping **description** and **imDb_id** column because that column not needed for the analysis

Now, Some of the columns which is needed. So, we can fill these columns
 both **age_certification** columns  and **seasons** columns have to exessive null values, **age_certification** is filled with mode and **season** with null values can be filled with '0' because movies are realse with a single id.

 **imDb_score** and **tmdb_score**  is replaced with median

 **imdb_votes** and **tmdb_popularity** is filled with '0'

````python
#drop description column
prime.drop('description', axis=1, inplace=True)
prime.drop('imdb_id', axis=1, inplace=True)
````

````python
#fill age certification with mode
prime['age_certification'].fillna(prime['age_certification'].mode()[0], inplace=True)

#fill season with 0
prime['seasons'].fillna(0, inplace=True)
````

````python
prime.isnull().sum()
````

````python
#filling null values with medium and zero
prime['imdb_score'].fillna(prime['imdb_score'].median(), inplace=True)
prime['tmdb_score'].fillna(prime['tmdb_score'].median(), inplace=True)
prime['imdb_votes'].fillna(0, inplace=True)
prime['tmdb_popularity'].fillna(0, inplace=True)
````

````python
#check the null values

prime.isnull().sum()
````

````python
#Replace charaters null values with unknown or N/A
prime['character'].fillna('N/A', inplace=True)
````

````python
print('shape:',prime.shape)
print(prime.isnull().sum())
#data is cleaned

print('Duplicated values:',prime.duplicated().sum())
````

````python
#drop the duplicates for better analysis
prime.drop_duplicates(inplace=True)
print('shape:',prime.shape)
````

#### Chart -16

````python
# Chart 16: Highest Rated Movies & Shows (IMDb Score)


#  Sort content by IMDb rating (top 15)
top_rated = (
    prime.groupby(["title", "type"], as_index=False)["imdb_score"]
    .max()
    .sort_values("imdb_score", ascending=False)
    .head(15)
)

print(top_rated)
# Step 2: Plot bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=top_rated, x="imdb_score", y="title", hue="type")

plt.title("Top 15 Highest Rated Movies & Shows on Amazon Prime", fontsize=14)
plt.xlabel("IMDb Rating")
plt.ylabel("Title")
plt.legend(title="Type")
plt.show()
````
